```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/log.csv', sep = ';')
df = df.replace(',', '.', regex=True)
df = df.astype(float)
df = df.rename(index={0: 'H2A', 1: 'H2B', 2: 'H3', 3: 'H4', 4: 'KDM3A'})
df
```

```python
f, ax = plt.subplots(figsize=(20,5))
colors = ['#000000','#1c001a','#290026','#40003b','#52024b','#b700cf','#ffffff']
ax = sns.heatmap(df, cmap = colors, annot = df)
ax.set_yticklabels(df.index.values)
plt.show()
```
